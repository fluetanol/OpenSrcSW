```python
#practice1

import matplotlib.pyplot as plt
import csv
f= open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)                    #<- header reading

date = header.index('날짜')
mintemp = header.index('최저기온(℃)') #<  index reading
maxtemp = header.index('최고기온(℃)')

maxtable=[]
mintable=[]

for row in data:                       #<-data reading
    splitdate = row[date].split('-');
    
    if splitdate[0] > '2010' and splitdate[1] == '04' and splitdate[2]=='15':
        maxtable.append(float(row[maxtemp]))
        mintable.append(float(row[mintemp]))
f.close()

plt.title('Birthday Max, Min Graph')
plt.plot(range(2011,2022),maxtable,'r',label='max')
plt.plot(range(2011,2022),mintable,'b',label='min')
plt.legend()
plt.show()
```

```python
f= open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)                    #<- header reading

date = header.index('날짜')
mintemp = header.index('최저기온(℃)') #<  index reading
maxtemp = header.index('최고기온(℃)')

Augtable=[]
Jantable=[]

for row in data:                       #<-data reading
    splitdate = row[date].split('-');
    if splitdate[0] > '2010':
        if splitdate[1] == '08':
            Augtable.append(float(row[maxtemp]))
        elif splitdate[1] == '01':
            Jantable.append(float(row[mintemp]))  
    
f.close()

plt.title('Histogram')
plt.hist(Augtable,bins=100,color='r',label='Aug')  #bins는 쪼개는 수준
plt.hist(Jantable,bins=100,color='b',label='Jan')  #bins는 쪼개는 수준
plt.legend()
plt.show()
```
