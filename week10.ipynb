```python
#실습 4 방법 1

phonebookname={}
phonebooknum={}
for i in range(10):
    a = input("이름 전화번호>")
    a = a.split(" ")
    phonebookname[a[0]]=a[1]
    phonebooknum[a[1]]=a[0]

while 1:
    k=int(input("1. 이름검색 2. 전화번호검색 3.종료 >"));
    if k==1:
        ans = input("name> ")
        print(phonebookname[ans])
    
    elif k==2:
        ans = input("telnum> ")
        print(phonebooknum[ans])
    
    elif k==3:
        print("종료");
        break;
        
    else:
        print("잘못된 입력");

```

```python
#실습 4 방법 2
phonebook=[]

count=0;
for i in range(2):
    a = input("이름 전화번호>")
    a = a.split(" ")
    phonebook.append(a[0])
    phonebook.append(a[1])

    
while 1:
    k=int(input("1. 이름검색 2. 전화번호검색 3.종료 >"));
    check = 1;
    
    if k==1:
        ans = input("name> ")
        for i in range(0,len(phonebook),2):
            if ans == phonebook[i]:
                print(phonebook[i+1])
                check=0
                break;
        if check==1:
            print("검색 실패")
    
    elif k==2:
        ans = input("telnum> ")
        for i in range(1,len(phonebook),2):
            if ans == phonebook[i]:
                print(phonebook[i-1])
                check=0
                break;
        if check==1:
            print("검색 실패")
    
    elif k==3:
        print("종료");
        break;
        
    else:
        print("잘못된 입력");
```
