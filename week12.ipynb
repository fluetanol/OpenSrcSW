```python
#실습 1 

import csv

f= open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)                    #<- header reading

date = header.index('날짜')
mintemp = header.index('최저기온(℃)') #<  index reading
maxtemp = header.index('최고기온(℃)')

staticdate = ['','','']
static = [-999,999,-999]

for row in data:                       #<-data reading
    if(row[mintemp]==''or row[maxtemp]==''):
        row[maxtemp]=-999
        row[mintemp]=999
    row[maxtemp] = float(row[maxtemp])
    row[mintemp] = float(row[mintemp])
    
    if static[0]<row[maxtemp]-row[mintemp]:
        static[0]=row[maxtemp]-row[mintemp]
        staticdate[0] = row[date]
        
    if static[1]>row[mintemp]:
        static[1]=row[mintemp]
        staticdate[1] = row[date]
        
    if static[2]<row[maxtemp]:
        static[2]=row[maxtemp]
        staticdate[2] = row[date]
    
    
    
    
f.close()

print('일교차가 가장 큰 날: ',static[0],'도, 날짜: ',staticdate[0])
print('최저기온이 가장 낮은 날',static[1],'도, 날짜: ',staticdate[1])
print('최고기온이 가장 높은 날',static[2],'도, 날짜: ',staticdate[2])
```

```python
#실습 2

f= open('subwayfee.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)     

linename = header.index('호선명')
subwayname = header.index('지하철역') #<  index reading
enter = header.index('승차승객수')
outer = header.index('하차승객수') 

uselist = [[['',-999,'',99999999],['',-999,'',99999999],['',-999,'',99999999],'']];
index=0;


for row in data:
    row[enter]=row[enter].replace(',','')
    row[outer]=row[outer].replace(',','')
    passenger=float(row[enter])+float(row[outer])
    
    if uselist[index][3]!=row[linename]:
        if len(uselist)!=1:
            index+=1;
        uselist.append([['',-999,'',99999999],['',-999,'',99999999],['',-999,'',99999999],''])
        uselist[index][3]=row[linename]
        
    if uselist[index][0][1]<passenger:
        uselist[index][2][1]=uselist[index][1][1]
        uselist[index][1][1]=uselist[index][0][1]
        uselist[index][0][1]=passenger
        uselist[index][2][0]=uselist[index][1][0]
        uselist[index][1][0]=uselist[index][0][0]
        uselist[index][0][0]=row[subwayname]
        
    elif uselist[index][1][1]<passenger:
        uselist[index][2][1]=uselist[index][1][1]
        uselist[index][1][1]=passenger
        uselist[index][2][0]=uselist[index][1][0]
        uselist[index][1][0]=row[subwayname]
        
    elif uselist[index][2][1]<passenger:
        uselist[index][2][1]=passenger
        uselist[index][2][0]=row[subwayname]
          
            
    if uselist[index][0][3]>passenger:
        uselist[index][2][3]=uselist[index][1][3]
        uselist[index][1][3]=uselist[index][0][3]
        uselist[index][0][3]=passenger
        uselist[index][2][2]=uselist[index][1][2]
        uselist[index][1][2]=uselist[index][0][2]
        uselist[index][0][2]=row[subwayname]
       
    elif uselist[index][1][3]>passenger:
        uselist[index][2][3]=uselist[index][1][3]
        uselist[index][1][3]=passenger
        uselist[index][2][2]=uselist[index][1][2]
        uselist[index][1][2]=row[subwayname]
        
    elif uselist[index][2][3]>passenger:
        uselist[index][2][3]=passenger
        uselist[index][2][2]=row[subwayname]
    
    
for i in range(0,len(uselist)-1):
    print("호선명: ",uselist[i][3])
    print("이용객이 가장 많은 역 순위")
    print("1. ",uselist[i][0][0],", 승객수: ",uselist[i][0][1])
    print("2. ",uselist[i][1][0],", 승객수: ",uselist[i][1][1])
    print("3. ",uselist[i][2][0],", 승객수: ",uselist[i][2][1])
    print("이용객이 가장 적은 역 순위")
    print("1. ",uselist[i][0][2],", 승객수: ",uselist[i][0][3])
    print("2. ",uselist[i][1][2],", 승객수: ",uselist[i][1][3])
    print("3. ",uselist[i][2][2],", 승객수: ",uselist[i][2][3])
    print("\n--------------------------------\n")
    
    
```
