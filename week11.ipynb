```python
#file 1 -> file2복사 실습 6

while 1:
    inputs = input("cp 파일명1 파일명2> ");
    filename =inputs.split(" ");
    if filename[0]!="cp":
        print("cp명령어가 아닙니다.");
    else:
        break;
    
fr=open(filename[1],'r');
fw=open(filename[2],'w');
lines = fr.readlines();

for line in lines:
    fw.write(line);

    
fr.close();
fw.close();
```

```python
#wc명령어를 python으로 구현, file의 라인 수, 단어 수를 구하여 출력함. 실습 7

inputs = input("wc 파일명 입력> ");

filename =inputs.split(" ");

if filename[0]!="wc":
    print("wc명령어가 아닙니다.");
    
fr=open(filename[1],'r');
lines = fr.readlines();

linecnt=0;
wordcnt=0;

for line in lines:
    temp = line.split(' ');
    wordcnt = wordcnt + len(temp);
    linecnt+=1;

print("line: ",linecnt);
print("word: ",wordcnt);
```

```python
import import_ipynb
from class_stss import Stackclass


stack = Stackclass();
strs = input("계산식 입력: ");
calcstr = strs.split(" ");
strs = strs.replace(" ","");

i=0;
temp = 0;
lencon=0;
lock=0;
lock2=0;

print("len: ",len(calcstr));

while i != len(calcstr):
    print("now str: ",calcstr[i],"lock: ",lock,"lock2: ",lock2);

    if lock2==0 and (calcstr[i]=="+"or calcstr[i]=="-"):
            stack.opradd(calcstr[i]);
        
    elif lock2==0 and (calcstr[i]=="*"or calcstr[i]=="/"):
        stack.opradd(calcstr[i]);
        if lock==1:
            i=i+1;
            lock-=1;
            continue;
        lock-=1;
  
    elif lock2==0:
        lencon+=1;
        stack.numadd(int(calcstr[i]));
        if i!=len(calcstr)-1 and (calcstr[i+1]=="*"or calcstr[i+1]=="/"):
            lock+=1;
        
    if lock==0 and lencon>=2: 
        tempnum1 = int(stack.numpop());
        tempopr = stack.oprpop();
        tempnum2 = int(stack.numpop());
        if tempopr =="+":
            stack.numadd(tempnum1+tempnum2);
        elif tempopr =="-":
            stack.numadd(tempnum2-tempnum1);
        elif tempopr =="*":
            stack.numadd(tempnum1*tempnum2);
        elif tempopr =="/":
            stack.numadd(tempnum2/tempnum1);
        lencon=lencon-1;
        
        if lock2 !=0:
            i=temp;
            lock2=0;

        if lencon>=2:
            print("싶패");
            temp = i;
            i=i-1;
            lock2=1;
            continue;
            
    i=i+1;
            

print("계산식 출력",strs,"=",stack.numpop());        
```
